```python
import morphanalysis

import skimage
import matplotlib.pyplot as plt
```

```python
def plot(image):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.imshow(image)
    ax.set_axis_off()
    plt.show()
```

```python
cell_image = skimage.io.imread("Datasets/CTRLvsSTAB/control/MAX_SW MICE 1_SEC 1_GFAP_RIGHT_CA1 _CONTRA_1-2_2.tif")
```

```python
astrocyte = morphanalysis.Cell(cell_image, image_type='confocal')
```

```python
plot(astrocyte.cleaned_image)
```

```python
astrocyte.surface_area()
```
